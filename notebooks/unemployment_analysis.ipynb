```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/unemployment.csv")

# Preview
print(df)

# Plot national unemployment rate
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Quarter", y="UnemploymentRate", palette="mako")
plt.title("South Africa Unemployment Rate")
plt.ylabel("Unemployment Rate (%)")
plt.savefig("../images/unemployment_rate.png")
plt.show()

# Plot youth unemployment
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x="Quarter", y="YouthUnemploymentRate", marker="o", color="red")
plt.title("Youth Unemployment Rate (15–24)")
plt.ylabel("Rate (%)")
plt.savefig("../images/youth_unemployment.png")
plt.show()
```
